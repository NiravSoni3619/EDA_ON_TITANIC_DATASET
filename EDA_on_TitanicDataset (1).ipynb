```python
# create new column family size 
df['Parch'] = df['Parch'].astype(int)
df['SibSp'] = df['SibSp'].astype(int)
df['Family Size' ] = df['Parch'] + df['SibSp']

```

```python
def family_type(number):
    if number == 0 :
        return "Alone"
    elif number >= 1 & number <=4 :
        return "medium"
    return "Large"

df['Family_type'] = df['Family Size'].apply(family_type)
df.sample(5)
```

```python
df.drop( columns = ['SibSp' , 'Parch' , 'Family Size'], inplace= True)
```

```python
crosstab = pd.crosstab(df['Family_type'], df['Survived']).apply(lambda r : round((r/r.sum())*100, 1), axis = 1)
print(crosstab)
sns.countplot(x = 'Family_type' , hue = 'Survived' , data = df )
```

```python

# Removing outliers in Age , As it is in normal distribution so using formula mean - 3SD to mean + 3SD
# Calculate mean and standard deviation
mean_age = df['Age'].mean()
std_age = df['Age'].std()

# Define bounds for acceptable values (e.g., within 3 standard deviations)
lower_bound = mean_age - 3 * std_age
upper_bound = mean_age + 3 * std_age

# Filter out outliers
df_no_outliers = df[(df['Age'] >= lower_bound) & (df['Age'] <= upper_bound)]

df = df_no_outliers
df.shape
```

```python
# removing outliers in Fare
# Calculate Q1, Q3, and IQR
Q1 = df['Fare'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)
IQR = Q3 - Q1

# Define lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df = df[(df['Fare'] > lower_bound) & (df['Fare']<upper_bound)]

df.shape
```

```python
# one hot encoding 
df = pd.get_dummies(df, columns=['Sex', 'Embarked', 'Pclass', 'Family_type'], drop_first=True, dtype=int)
df.sample(5)
```
